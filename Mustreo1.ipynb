# How positives examples are sampling?

Our first sampling approach was stratified sampling using diagnosing, sex, age and anatomic location, but our positives examples are very few, and use stratified sampling with anatomic location it would be very unpractical. Hence we decided to use the first, second and thirth random variables to sample positives images, however in order to get more information from positives cases we are going to get more samples than negative class.

# But we have an unbalanced dataset, so we analize the dataset using a strata given by outputs labels

# Setting sample function that does stratified sampling using sex, age and localization

Creating a function that does stratified sampling with target, sex and age stratas

```python
def sample(df=train_df,sex="male", target=1, sampling_rate=0.15):
    #This function does stratified sampling over a df pandas DataFrame using target, sex and age stratas
    
    #Parameters
    #df: pandas DataFrame
    #sex: String value "male" or "female" in sex column from df
    #target: target Integer value 0 or 1 to sample from df
    #sampling_rate: float value between 0 or 1
    
    #Returns
    #sample_df: pandas DataFrame sampled from df
    
    #Only rows with the specified target 0 or 1
    df_target = df.query('target==@target')
    
    #Only rows with the sex specified
    df_sex=df_target.query('sex==@sex')
    
    ''' We have 15 differents ages, the minimum age is 20 and the maximum is 90 
    Hence we proppose the following stratas:
    Untill 29 years
    30 to 39 years
    40 to 49 years
    50 to 59 years
    60 to 69 years
    70 to 79 years
    80 or more'''
    
    #Setting stratas by age
    s1=df_sex.query('age_approx<=29')
    s2=df_sex.query('30<=age_approx<=39')
    s3=df_sex.query('40<=age_approx<=49')
    s4=df_sex.query('50<=age_approx<=59')
    s5=df_sex.query('60<=age_approx<=69')
    s6=df_sex.query('70<=age_approx<=79')
    s7=df_sex.query('age_approx>=80')
    
    #Getting number of samples per strata with around sampling_rate percent of original strata
    n1=s1['target'].count()*sampling_rate
    n2=s2['target'].count()*sampling_rate
    n3=s3['target'].count()*sampling_rate
    n4=s4['target'].count()*sampling_rate
    n5=s5['target'].count()*sampling_rate
    n6=s6['target'].count()*sampling_rate
    n7=s7['target'].count()*sampling_rate
    
    #Rounding and turning results into int64
    n1=(np.around(n1)).astype('int64') 
    n2=(np.around(n2)).astype('int64') 
    n3=(np.around(n3)).astype('int64') 
    n4=(np.around(n4)).astype('int64') 
    n5=(np.around(n5)).astype('int64') 
    n6=(np.around(n6)).astype('int64') 
    n7=(np.around(n7)).astype('int64') 
    
    #Now randomly looking for the examples per strata
    f1=s1.sample(n=n1)
    f2=s2.sample(n=n2)
    f3=s3.sample(n=n3)
    f4=s4.sample(n=n4)
    f5=s5.sample(n=n5)
    f6=s6.sample(n=n6)
    f7=s7.sample(n=n7)
    
    #Append results of sampling
    p=f1
    p=p.append(f2)
    p=p.append(f3)
    p=p.append(f4)
    p=p.append(f5)
    p=p.append(f6)
    p=p.append(f7)
    
    #Setting sample_df
    sample_df=p
    
    return sample_df
```
